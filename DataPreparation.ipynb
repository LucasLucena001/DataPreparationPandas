```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ks-projects-201801.csv")
```

```python
df.head()
```

```python
df = df.drop(columns =["name", "category", "goal", "pledged", "usd pledged"])
```

```python
df.head()
```

```python
df.info()
```

```python
df["launched"] = df["launched"].str.replace(" \d\d:\d\d:\d\d", "", regex = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["usd_goal_real"] = df["usd_goal_real"].astype("int64")
```

```python
df["usd_pledged_real"]= df["usd_pledged_real"].astype("int64")
```

```python
df.dtypes
```

```python
df["deadline"] = pd.to_datetime(df["deadline"],  format = '%Y-%m-%d')
df["launched"] = pd.to_datetime(df["launched"], format = "%Y-%m-%d")
```

```python
df["time_range"] = (df["deadline"] - df["launched"]).dt.days
```

```python
df_invest = pd.read_csv("invested.csv")
```

```python
df_invest.head()
```

```python
df_invest.sort_values(by = ["ID"])
```

```python
df_invest= pd.get_dummies(df_invest, columns = ["backedLocation"])
```

```python
df_invest.sort_values(by = ["ID"])
```

```python
df_invest = df_invest.groupby(by = ["ID"]).agg({"age": "mean", "backedLocation_BR": "sum", "backedLocation_GBK": "sum", "backedLocation_US": "sum"})
```

```python
df_invest.head()
```

```python
df = df.merge(df_invest, how ="right", on= "ID")
```
